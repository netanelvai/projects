```python
pg_color={'Adelie':'#FFFFCC',
          'Chinstrap':'#1C81A5',
          'Gentoo':'#65A51C'
        }
```

```python
sns.set_style('darkgrid')
sns.barplot(x='species',y='body_mass_g',data=pg,palette=pg_color)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.barplot(x='species',y='body_mass_g',data=pg,palette=pg_color)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.set_style('whitegrid')
sns.barplot(x='species',y='body_mass_g',data=pg,palette=pg_color)
plt.show()
```

```python
#important 
pg.groupby('species').body_mass_g.describe()
```

```python
plt.figure(figsize=(8,5))
sns.set_style('whitegrid')
sns.barplot(x='species',y='body_mass_g',data=pg,palette=pg_color,estimator=np.min)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.set_style('whitegrid')
sns.barplot(x='species',y='body_mass_g',data=pg,palette=pg_color,estimator=np.max)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.set_style('whitegrid')
sns.barplot(x='species',y='body_mass_g',data=pg,hue='sex')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.set_style('whitegrid')
sns.barplot(x='species',y='body_mass_g',data=pg,hue='sex',order=['Gentoo','Chinstrap','Adelie'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.set_style('whitegrid')
sns.barplot(x='species',y='body_mass_g',data=pg,hue='sex',order=['Gentoo','Chinstrap','Adelie']
           ,hue_order=['Female','Male'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.set_style('whitegrid')
sns.barplot(x='species',y='body_mass_g',data=pg,hue='sex',order=['Gentoo','Chinstrap','Adelie']
           ,hue_order=['Female','Male'])
plt.xlabel('types',fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.set_style('whitegrid')
sns.barplot(x='species',y='body_mass_g',data=pg,hue='sex',order=['Gentoo','Chinstrap','Adelie']
           ,hue_order=['Female','Male'])
plt.xlabel('types',fontsize=14)
plt.ylabel('weight',fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.set_style('whitegrid')
sns.barplot(x='species',y='body_mass_g',data=pg,hue='sex',order=['Gentoo','Chinstrap','Adelie']
           ,hue_order=['Female','Male'])
plt.title('types/avg weight', color='green',fontsize=20)
plt.xlabel('types',fontsize=14)
plt.ylabel('weight',fontsize=12)
plt.show()
```
